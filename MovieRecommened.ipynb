```python
import pandas as pd
import numpy as np
```

```python
try:
    data=pd.read_csv('/content/dataset.csv')
except pd.errors.ParserError as e:
    print(f"Error reading CSV: {e}")
    # If the error persists, try specifying the 'error_bad_lines' parameter
    data = pd.read_csv('/content/dataset.csv', error_bad_lines=False)
    print("CSV read with 'error_bad_lines=False', skipping problematic lines.")
data.tail()
```

```python
data.info()
```

```python
data['tags']=data['genre']+data['overview']
data.head()
```

```python
new_data=data[['id','title','overview','genre']]
new_data=data.drop(columns=['overview','genre'])
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(max_features=10000,stop_words='english')
vec=vectorizer.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vec.shape
```

```python
#find simliarities between them to if see one another all are rcommented
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vec)
```

```python
similarity
```

```python
new_data[new_data['title']=='The Lego Movie'].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].title)
```

```python
def rcommented(data):
  index=new_data[new_data['title']==data].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
Movies_title=str(input("Enter the movie name:"))
rcommented(Movies_title)
```
