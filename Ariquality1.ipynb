```python
#Forecasting with FB Prophet algorithm
#FB Prophet algorithm only contain two coloums
#one with dateand timestamp another predict data
#date and time stamp like YYYY-MM-DD HH:MM:SS or YYYY-MM-DD
```

```python
#convert the date and time stamp according to FB Prophet
#convert the date and time stamp according to FB Prophet
date_info = pd.to_datetime(air_quality_data['Date'], dayfirst=True).dt.strftime('%Y-%m-%d')
print(date_info)
```

There are two types of data forecasting
1 Boston house pricing (with no date and time-use Linear Regression)
2 Forecasting with date and time

---






```python
#convert the '.' into ':'
time_info=air_quality_data['Time']
time_info=time_info.apply(lambda x:x.replace('.',':'))
#THIS TO CONVERT THE TIMESTAMP BUT IT DOESN'T WANT
#time_info = pd.to_datetime(air_quality_data['Time'], dayfirst=True).dt.strftime('%H:%M:%S')
print(time_info)
```

```python
print(type(time_info))
print(type(date_info))
#the are two d_type in pandas one dataframe and sereies
#dataframe has multiple row and coloumn and dynamic
#series has single row and single coloumn and static
```

```python
#make time and date use pandas dataframe for keep only two cloums
date_time=pd.concat([date_info,time_info],axis=1)
print(date_time)
```

```python
#combining date and time coloumn and name the coloumn ds according to FB Prophet docs
date_time['ds']=date_time['Date']+' '+date_time['Time']
date_time.head()
```

```python
data=pd.DataFrame()
```

```python
#convert ds in to date_time date type
data['ds']=pd.to_datetime(date_time['ds'])
data.info()
```

```python
data.head()
```

```python
data['y']=air_quality_data['C6H6(GT)']
data.head()
```

```python
from prophet import Prophet
```

```python
#traing prophet model
model=Prophet(yearly_seasonality=True)
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
```

```python
fig2=model.plot_components(forecast)
```
